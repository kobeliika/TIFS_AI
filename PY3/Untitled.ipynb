```python
words = []

while True:
    word = input("Enter an word:")
    if word == '':
        break
    else:
        words.append(word)
unique_words = []
for word in words:
    if word not in unique_words:
        unique_words.append(word)
for word in unique_words:
    print(word)
```

```python
lst = []
while True:
    element = input("Enter the element:")
    if element == '':
        break
    else:
        lst.append(element)
for i in range(len(lst)-1):
    print(lst[i], end='')
    if lst[i+1] == lst[-1]:
        print(" and ", lst[i+1], end='')
    else:
        print(", ", end='')
```
